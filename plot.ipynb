```python
import numpy as np
import matplotlib.pyplot as plt

result = {}
with open('result.csv') as file:
    for line in file:
        line = line.strip()
        result[float(line.split(' ')[0])] = complex(line.split(' ')[1]).real
                
plt.figure(figsize=(12, 6))
plt.plot(result.keys(), result.values(), label="Real Part of A(ω)")
plt.title('Fourier Transform of the Spin Evolution')
plt.xlabel('Frequency (ω)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
